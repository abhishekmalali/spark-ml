```python
#import numpy as np
import matplotlib.pyplot as plt
#import scipy as sc
import autograd.scipy as sc  # Thinly-wrapped scipy
import autograd.numpy as np  # Thinly-wrapped numpy
from autograd import grad
from sklearn.linear_model import LogisticRegression
import os
import pandas as pd
import itertools
from parse import *
import shutil
import seaborn as sns
sns.set_style('whitegrid')
```

```python
df = pd.read_csv('../cache/musicdata.csv', header=None)
df.columns = ['uid', 'aid', 'rating']
```

```python
# get ratings for the beatles
beatles_id = 1002217
dftouse = (df[df['aid'] == beatles_id])[['uid', 'rating']]
dftouse[dftouse['rating'] == 255] = 0
dftouse[['rating']] = np.floor(dftouse['rating'] / 20)
dftouse = dftouse.head(2000)
```

```python
%matplotlib inline
plt.scatter(dftouse['uid'], dftouse['rating'], alpha=.2)
ax = plt.gca()
ax.xaxis.set_major_formatter(plt.NullFormatter())
plt.ylim([0.5, 5.5])
plt.xlim([0, dftouse['uid'].max()])
plt.ylabel('rating')
plt.title('The Beatles rated according to Yahoo users (2004)')
plt.savefig('beatles.png', dpi=150)
```
